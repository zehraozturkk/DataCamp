## The data

| Column | Description |

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

 #Start coding!

#I read netflixdata.csv
netflix_df = pd.read_csv("netflix_data.csv",index_col=0)

#we take just movie type
netflix_subset = netflix_df[netflix_df["type"]=="Movie"]

#we filter movies by "title", "country", "genre", "release_year", "duration" and create new data frame called netflix_movies
netflix_movies = netflix_subset[["title", "country", "genre", "release_year", "duration" ]]

#nm = netflix_subset.loc[:,["title", "country", "genre", "release_year", "duration" ]]

#I take some movie that duration is less than 60.
short_movies = netflix_movies[netflix_movies["duration"]<60]

colors = []
for label,row in netflix_movies.iterrows():
    if row["genre"]=="Children":
        colors.append("yellow")
    elif row["genre"]=="Documentaries":
        colors.append("green")
    elif row["genre"]=="Stand-up":
        colors.append("blue")
    else:
        colors.append("gray")
    
fig =  plt.figure(figsize=(12,8))
plt.scatter(netflix_movies["release_year"], netflix_movies["duration"], c = colors)

plt.title("Movie Duration by Year of Release")
plt.xlabel("Release year")
plt.ylabel("Duration (min)")

answer = "maybe"



plt.show()
```
